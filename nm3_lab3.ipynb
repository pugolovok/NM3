```python
L = 12%13 + 1 # метров
R = 1 * 0.0254 # метров
mu = 0.934 # Па*с
p = 5 * 1000 # Па
```

```python
def pryamoy_hod(L, mu, R, p, N):
    h = R / N
    D = (p * h ** 2) / (mu * L)
    A = 1
    B = 1
    C = -2
    alfa = [1, 1]
    beta = [0, 0]
    for i in range(2, N + 1, 1):
        alfa.append(-B / (A * alfa[-1] + C))
        beta.append((D - A * beta[-1]) / (A * alfa[-1] + C))
    return alfa, beta
```

```python
def obratniy_hod(alfa, beta, N):
    u = [0]
    for i in range(1, N+1):
        u.append(alfa[len(alfa)-i] * u[len(u)-1] + beta[len(beta)-i])
    u.pop(0)
    u = u[::-1]
    return u
```

```python
N = [5, 10, 15, 20]
u_i =  []
for n in N:
    alfa, beta = pryamoy_hod(L, mu, R, p*(-1), n)
    u_i.append(obratniy_hod(alfa, beta, n))
for i in range(len(u_i)):
    print(f"N = {N[i]}")
    delta_x = R / N[i]
    from_center = 0.0
    for j in u_i[i]:
        print("Расстояние от центра:", round(from_center, 5), "; скорость:",  round(j, 5))
        from_center += delta_x
    print()
```
